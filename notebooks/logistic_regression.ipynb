# scikit learn in 3 lines of code except actually it's more like 20


First, we'll generate a dataset. It has two features called `panda` and `elephant`, and we generate a binary output target variable `target`. 

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.DataFrame({
    'panda': np.random.normal(0, 1, 100000),
    'elephant': np.random.normal(0, 1, 100000)
})
x = - 1/3 * (dataset['panda'] +  dataset['elephant'])
transformed = 1 / (1 + np.exp(-1 * x))
dataset['target'] = np.random.uniform(0,1, 100000) < transformed
dataset.target.value_counts()
```

Now that we have fake data, we can build a classifier!


```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
cls = LogisticRegression()
features = dataset[['elephant', 'panda']]
target = dataset['target']
features_train, features_test, target_train, target_test = train_test_split(features, target)
cls.fit(features_train, target_train)
predictions = cls.predict(features_test)
print confusion_matrix(predictions, target_test)
```


So we have a classifier which is classifying 8953 + 9030 things correctly, and 3508 + 3500 things wrong. That's very good! That was legitimately something like 3 lines of code -- most of the work was actually setting up the fake dataset.

```python
pd.set_option('display.mpl_style', 'default')
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
scores = cls.predict_proba(features_test).transpose()[1]
fpr, tpr, thresholds = roc_curve(target_test, scores)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.plot(fpr, tpr)
```

```python
trues = dataset[['elephant', 'panda']][target][:500]
falses = dataset[['elephant', 'panda']][~target][:500]
ax = plt.axes()
trues.plot(x='elephant', y='panda', kind='scatter', ax=ax, alpha=0.6, label='True')
falses.plot(x='elephant', y='panda', kind='scatter', color='orange', ax=ax, alpha=0.6, label='False')
plt.legend()
```

```python
trues['elephant'] -= 1
trues['panda'] -= 1
ax = plt.axes()
trues.plot(x='elephant', y='panda', kind='scatter', ax=ax, alpha=0.6, label='True')
falses.plot(x='elephant', y='panda', kind='scatter', color='orange', ax=ax, alpha=0.6, label='False')
plt.legend()
```

```python
test_df = pd.DataFrame(features_test, columns=['panda', 'elephant'])
trues = test_df[['elephant', 'panda']][predictions][:500]
falses = test_df[['elephant', 'panda']][~predictions][:500]
ax = plt.axes()
trues.plot(x='elephant', y='panda', kind='scatter', ax=ax, alpha=0.6, label='True')
falses.plot(x='elephant', y='panda', kind='scatter', color='orange', ax=ax, alpha=0.6, label='False')
plt.legend()
```

```python
rounded_scores = np.floor(scores * 100)
```

```python
actual_proportions = np.zeros(100)
for i in xrange(100):
    values = pd.Series(target_test[rounded_scores == i]).value_counts()
    values = values.reindex([True, False]).fillna(0)
    actual_proportions[i] = values.ix[True] * 1.0 / (values.ix[True] + values.ix[False])
```

```python
plt.plot(np.arange(0, 1, 0.01), actual_proportions)
plt.xlabel('Predicted probability')
plt.ylabel('Actual proportion of Trues')
```
